```python
df = pd.read_excel("check_files_final.xlsx", dtype='str')

df['possible_dates'] = df['possible_dates'] \
                            .str.replace("[", "").str.replace("]", "") \
                            .str.replace("'", "") \
                            .str.split(", ") \
                            .apply(lambda x: sorted(list(set(x))))
df['first_dt'] = pd.to_datetime(df['possible_dates'].str[0], format='%Y%m%d')
df['file_dt'] = pd.to_datetime(df['filename'].str.split('.').str[0] \
                                                   .str.split('_').str[0] \
                                                   .str.split(' ').str[0] \
                                                   .str.split('-').str[0], 
                               format='%Y%m%d')
df['diff_days'] = df['first_dt'] - df['file_dt']
df
```

```python
df = pd.read_excel("checked_results.xlsx", dtype='str')
df.loc[df['DateTimeOriginal'] == df['FinalDateTime'], 'meta_change_flg'] = 'DateTimeOriginal'
df['possible_new_name'] = None
df['random_i'] = 1
df
```

### Get new name

```python
for i in range(len(df)):
    name_dt = df.loc[i, 'meta_change_flg']
    if name_dt != 'random':
        df.loc[i, 'possible_new_name'] = df.loc[i, name_dt][:19].replace(':', '').replace(' ', '_').replace('+03:00', '')

    else:
        df.loc[i, 'possible_new_name'] = df.loc[i, 'filename'].split('.')[0].split('_')[0].split(' ')[0].split('-')[0] + '_123456'
        random_i = None
        if '(' in df.loc[i, 'filename']:
            random_i = df.loc[i, 'filename'].split(')')[0].split('(')[1]
        df.loc[i, 'random_i'] = int(random_i) if random_i else 1

df = df.sort_values(['possible_new_name', 'random_i']).reset_index(drop=True)
df['cnt'] = df.groupby('possible_new_name')['possible_new_name'].transform('size')
df['cum_cnt'] = df.groupby('possible_new_name').cumcount() + 1

df['new_name'] = None
df.loc[0, 'new_name'] = df.loc[0, 'possible_new_name'][:8]
for i in range(1, len(df)):
    prev_dt = df.loc[i - 1, 'possible_new_name']
    cur_dt = df.loc[i, 'possible_new_name']    

    if prev_dt[:8] != cur_dt[:8]:
        df.loc[i, 'new_name'] = cur_dt[:8]
    elif prev_dt != cur_dt and len(prev_dt) != len(df.loc[i - 1, 'new_name']) and len(df.loc[i - 1, 'new_name']) <= 15:
        df.loc[i - 1, 'new_name'] = prev_dt
        df.loc[i, 'new_name'] = cur_dt
    elif prev_dt != cur_dt:
        df.loc[i, 'new_name'] = cur_dt
    else:
        len_postfix = len(str(df.loc[i, 'cnt']))
        df.loc[i - 1, 'new_name'] = f"{prev_dt}_{df.loc[i - 1, 'cum_cnt']:0{len_postfix}d}"
        df.loc[i, 'new_name'] = f"{cur_dt}_{df.loc[i, 'cum_cnt']:0{len_postfix}d}"

df['new_name'] = df['new_name'] + '.' + df['filename'].str.extract(r'\.([^.]+)$', expand=False).str.lower().replace('jpeg', 'jpg').fillna('')
df
```

### Get meta

```python
import subprocess
import json


def get_metadata(file_path):
    try:
        # Run exiftool command to get metadata in JSON format
        result = subprocess.run(
            [r'C:\Users\crisc\Downloads\exiftool-13.18_64\exiftool.exe', '-j', file_path],
            stdout=subprocess.PIPE,
            stderr=subprocess.PIPE,
            text=True
        )

        # Check if the command was successful
        if result.returncode == 0:
            # Parse the JSON output
            metadata = json.loads(result.stdout)
            # print(metadata[0])
            return metadata[0]  # Return the first (and only) dictionary in the list
        else:
            print(f"Error: {result.stderr}")
            return None
    except Exception as e:
        print(f"An error occurred: {e}")
        return None


def change_metadata(file_path, key, value):
    """
    Change metadata of a file using Exiftool.

    :param file_path: Path to the file whose metadata you want to change.
    :param metadata: A dictionary containing the metadata tags and their new values.
    """

    exiftool_path = r'C:\Users\crisc\Downloads\exiftool-13.18_64\exiftool.exe'
    
    # Construct the Exiftool command
    command = [exiftool_path]
    command.append('-overwrite_original')
    
    # Add metadata changes to the command
    command.append(f'-{key}={value}')
    
    # Add the file path to the command
    command.append(file_path)
    
    # Run the command
    result = subprocess.run(command, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
    
    # Check if the command was successful
    if result.returncode != 0:
        print("Failed to change metadata.")
        print(result.stderr.decode())
```

```python
# need to change meta for mov filecreatedt if there is no '+03:00' and fcd != cd
# need to change +04:00 so get custom check
# need to add to fcd / cd +03:00
# to the samp datetime add 1 second

# for videos mov - should be FileCreateDate as datetime with timezone and CreationDate as datetime with timezone
# for other types - should be FileCreateDate as datetime with timezone and DateTimeOriginal + OffsetTimeOriginal

from datetime import datetime, timedelta

photo_dir = os.path.join(os.getcwd(), 'Life')
datetime_format = "%Y:%m:%d %H:%M:%S"
df['has_meta_change'] = False
df['NewFileCreateDate'] = df['FileCreateDate']
df['NewCreationDate'] = None
df['NewDateTimeOriginal'] = None
for i in range(len(df)):
    file_path = os.path.join(photo_dir, df.loc[i, 'new_name'])
    metadata = get_metadata(file_path)
    
    offset_time_original = metadata["OffsetTimeOriginal"] if metadata and "OffsetTimeOriginal" in metadata else None
    file_create_dt = metadata["FileCreateDate"][:19] if metadata and "FileCreateDate" in metadata else None
    
    file_ext = file_path.rsplit('.', maxsplit=1)[1].lower()
    name_dt = df.loc[i, 'meta_change_flg']
    right_time_another_timezone = None
    
    if not pd.isna(df.loc[i, 'NewOffsetTimeOriginal']) and offset_time_original != df.loc[i, 'NewOffsetTimeOriginal']:
        df.loc[i, 'has_meta_change'] = True
        offset_time_original = df.loc[i, 'NewOffsetTimeOriginal']

    if file_ext == 'mov':
        right_dt = df.loc[i, name_dt]
        create_dt = metadata["CreationDate"][:19] if "CreationDate" in metadata else None
        
        if '+' not in right_dt:
            right_dt += time_offset
            
        if '+04:00' in right_dt:
            right_time_another_timezone = right_dt
            right_dt = datetime.strptime(right_dt[:19], datetime_format)
            right_dt = right_dt - timedelta(hours=1)
            right_dt = right_dt.strftime(datetime_format) + '+03:00' # hard code
            
        if not right_time_another_timezone and right_dt.split('+')[0] != create_dt:
            df.loc[i, 'has_meta_change'] = True
            df.loc[i, 'NewCreationDate'] = right_dt
            change_metadata(file_path, 'CreationDate', right_dt)
        elif right_time_another_timezone and right_time_another_timezone.split('+')[0] != create_dt:
            df.loc[i, 'has_meta_change'] = True
            df.loc[i, 'NewCreationDate'] = right_time_another_timezone
            change_metadata(file_path, 'CreationDate', right_time_another_timezone)
        if right_dt.split('+')[0] != file_create_dt:
            df.loc[i, 'has_meta_change'] = True
            df.loc[i, 'NewFileCreateDate'] = right_dt
            change_metadata(file_path, 'FileCreateDate', right_dt)
    elif name_dt == 'random':
        right_dt = datetime.strptime(df.loc[i, 'possible_new_name'], "%Y%m%d_%H%M%S")
        right_dt = right_dt + timedelta(seconds=int(df.loc[i, 'cum_cnt']) - 1)
        right_dt = right_dt.strftime(datetime_format) + '+03:00' # hard code
        
        datetime_original = metadata["DateTimeOriginal"] if metadata and "DateTimeOriginal" in metadata else None

        if '+' not in right_dt:
            right_dt += time_offset
            
        if '+04:00' in right_dt:
            right_time_another_timezone = right_dt
            right_dt = datetime.strptime(right_dt[:19], datetime_format)
            right_dt = right_dt - timedelta(hours=1)
            right_dt = right_dt.strftime(datetime_format) + '+03:00' # hard code
            
        if not right_time_another_timezone and right_dt.split('+')[0] != datetime_original:
            df.loc[i, 'has_meta_change'] = True
            df.loc[i, 'NewDateTimeOriginal'] = right_dt.split('+')[0]
            change_metadata(file_path, 'DateTimeOriginal', right_dt.split('+')[0])
        elif right_time_another_timezone and right_time_another_timezone.split('+')[0] != datetime_original:
            df.loc[i, 'has_meta_change'] = True
            df.loc[i, 'NewDateTimeOriginal'] = right_time_another_timezone.split('+')[0]
            change_metadata(file_path, 'DateTimeOriginal', right_time_another_timezone.split('+')[0])

        if right_dt.split('+')[0] != file_create_dt:
            df.loc[i, 'has_meta_change'] = True
            df.loc[i, 'NewFileCreateDate'] = right_dt
            change_metadata(file_path, 'FileCreateDate', right_dt)
    else:
        right_dt = df.loc[i, name_dt]
        datetime_original = metadata["DateTimeOriginal"] if metadata and "DateTimeOriginal" in metadata else None

        if '+' not in right_dt:
            right_dt += time_offset
            
        if '+04:00' in right_dt:
            right_time_another_timezone = right_dt
            right_dt = datetime.strptime(right_dt[:19], datetime_format)
            right_dt = right_dt - timedelta(hours=1)
            right_dt = right_dt.strftime(datetime_format) + '+03:00' # hard code
            
        if not right_time_another_timezone and right_dt.split('+')[0] != datetime_original:
            df.loc[i, 'has_meta_change'] = True
            df.loc[i, 'NewDateTimeOriginal'] = right_dt.split('+')[0]
            change_metadata(file_path, 'DateTimeOriginal', right_dt.split('+')[0])
        elif right_time_another_timezone and right_time_another_timezone.split('+')[0] != datetime_original:
            df.loc[i, 'has_meta_change'] = True
            df.loc[i, 'NewDateTimeOriginal'] = right_time_another_timezone.split('+')[0]
            change_metadata(file_path, 'DateTimeOriginal', right_time_another_timezone.split('+')[0])
        if right_dt.split('+')[0] != file_create_dt:
            df.loc[i, 'has_meta_change'] = True
            df.loc[i, 'NewFileCreateDate'] = right_dt
            change_metadata(file_path, 'FileCreateDate', right_dt)

        print(file_path.rsplit('\\', maxsplit=1)[1], "DONE")
```
